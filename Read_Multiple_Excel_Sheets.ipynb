### Importing Libraries

### Read Data

```python
# Read All Sheets In The Excel File as Dictionary
excel_sheet = pd.read_excel("supply_data.xlsx", sheet_name = None)
print(type(excel_sheet))
excel_sheet
```

### Concat All Sheets Together

```python
# Create Empty df
df = pd.DataFrame()
```

```python
# Looping on all Sheets To Concat it with the df
for k, _ in excel_sheet.items():
    df = pd.concat([df, excel_sheet[k]], ignore_index=True)
```

```python
df.shape
```

### Replace Space From Columns With _

```python
df.columns = df.columns.str.replace(" ", "_")
df.columns
```
